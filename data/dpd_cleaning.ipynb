Ovevřeni souboru s ruiany

```python
import pandas as pd

df = pd.read_csv('dpd_getAll_country203_kod_obce.csv')

df.head(5)
```

```python
df.columns
```

kontrola sloupců a vyběr jen pár potřebných

```python
vybrane_sloupce = ['company', 'kod_obce', 'city', 'postcode', 'street', 'house_number', 'latitude', 'longitude' ]
df_vybrany = df[vybrane_sloupce]
```

```python
df_vybrany.head(5)
```

Přejmenování souborů

```python
nove_nazvy_sloupcu = {'company': 'name',
                      'kod_obce': 'city_code',
                      'city': 'city',
                      'postcode':'zip_code',
                      'street': 'street',
                      'house_number': 'house_NO',
                      'latitude': 'latitude',
                      'longitude' :'longitude' 
                      }
df_prejmenovany = df_vybrany.rename(columns = nove_nazvy_sloupcu)
```

```python
df_prejmenovany.head(5)
```

uložení do noveho csv

```python
df_prejmenovany.to_csv('clean_data_dpd.csv', index=False, encoding='utf-8')
```
